```python
first = ((d3.iloc[:,-1] - d3.iloc[:,-2]) -  (d3.iloc[:,-2] - d3.iloc[:,-3]))/ 2
```

```python
second = ((d3.iloc[:,-2] - d3.iloc[:,-3]) -  (d3.iloc[:,-3] - d3.iloc[:,-4]))/ 2
```

```python
third = ((d3.iloc[:,-3] - d3.iloc[:,-4]) -  (d3.iloc[:,-4] - d3.iloc[:,-5]))/ 2
```

```python
avgacc = (first+second+third)/3
```

```python
diff = d3.iloc[:,-1] - d3.iloc[:,-2]
```

```python
d3['slope'] = diff
d3['avgaccelration'] = avgacc
```
